# Face Detection

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Membuat objek CascadeClassifier dengan file haarcascade untuk deteksi wajah
cascade = cv2.CascadeClassifier('/content/drive/MyDrive/PCVK/Images/haarcascades/haarcascade_frontalface_alt.xml')
```

## Avifah

```python
# Membaca gambar 'avifah1.jpg', mengubahnya ke RGB, dan membuat salinan
avifah1 = cv2.imread('/content/drive/MyDrive/PCVK/ktm/avifah1.jpg')
avifah1_rgb = cv2.cvtColor(avifah1, cv2.COLOR_BGR2RGB)
avifah1_crop = avifah1_rgb[650:2790, 500:3900]
avifah1_fd = avifah1_crop.copy()

# Mendeteksi wajah pada gambar 'avifah1.jpg' dan menggambar kotak merah di sekitar wajah yang terdeteksi
fd_avifah1 = cascade.detectMultiScale(avifah1_fd, scaleFactor=2)
for(x,y,w,h) in fd_avifah1:
  cv2.rectangle(avifah1_fd, (x,y),(x+w,y+h), (255,0,0), 15)

# Menampilkan gambar 'avifah1.jpg' sebelum dan sesudah deteksi wajah
plt.figure(figsize = (10,10))
f, axarr = plt.subplots(1,2,figsize=(8,8))
plt.subplot(121)
plt.imshow(avifah1_rgb)
plt.title('Gambar Original')
plt.subplot(122)
plt.imshow(avifah1_fd)
plt.title('Face Detection')
plt.show()
```

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
import easyocr
from PIL import Image
```

## Avifah

```python
avifah1 = cv2.imread('/content/drive/MyDrive/PCVK/ktm/avifah1.jpg')
avifah1_rgb = cv2.cvtColor(avifah1, cv2.COLOR_BGR2RGB)
avifah1_crop = avifah1_rgb[650:2790, 500:3900]
avifah1_gray = cv2.cvtColor(avifah1_crop, cv2.COLOR_BGR2GRAY)
kernel = np.ones((5, 5), np.uint8)
avifah1_eroded = cv2.erode(avifah1_gray, kernel, iterations=1)

def ocr_id_card(image):
    reader = easyocr.Reader(['id'])
    image = np.array(image)
    result = reader.readtext(image)
    text = '\n'.join([item[1] for item in result])
    return text

plt.figure(figsize = (10,10))
f, axarr = plt.subplots(1,2,figsize=(8,8))
plt.subplot(121)
plt.title('Gambar Original')
plt.imshow(avifah1_rgb)
plt.subplot(122)
plt.title('Preprocessing')
plt.imshow(avifah1_eroded, cmap='gray')

print(ocr_id_card(avifah1_eroded))
```
